## Code to combine two halfes of CSV files

```python
from tqdm import tqdm
import pandas as pd

# Load both CSV files
csv_file1 = 'test_original.csv'
csv_file2 = 'test_FINAL_full.csv'
prediction_df = pd.read_csv(csv_file1)
result_df = pd.read_csv(csv_file2)

# Debug 
debug = False

# Iterate over all rows
for index, (row1, row2) in tqdm(enumerate(zip(prediction_df.itertuples(), result_df.itertuples())), total=len(prediction_df)):

    # Skipping initial rows
    # if index <= 65593: continue

    # Check the 'entity_value' column in both file
    prediction = getattr(row1, 'prediction')

    # Adding predictions  column
    row2['prediction'] = prediction

    # Testing
    # if index == 100: break

# Save the result to a new CSV file
result_df.to_csv('test_FINAL_full_allCols.csv', index=False)
```
